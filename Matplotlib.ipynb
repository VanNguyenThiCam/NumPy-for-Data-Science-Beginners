# Matplotlib với NumPy

```python
import numpy as np

import matplotlib.pyplot as plt
x = np.arange(1,11)
y = 2 * x + 5

plt.title('Đồ thị')
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.plot(x,y)
plt.show
```

```python
import numpy as np

import matplotlib.pyplot as plt
x = np.arange(1,11)
y = 2 * x + 5

plt.title('Đồ thị')
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.plot(x,y,'ob')
plt.show
```

```python
x = np.arange(0, 3*np.pi,0.1)
y = np.sin(x)
plt.title('Đồ thị hàm sinx')
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.plot(x,y)
plt.show
```
